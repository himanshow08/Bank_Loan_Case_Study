# Segmented Univariate Analysis:

## Gender:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Segmented Univariate Analysis - Stacked Bar Chart
def segmented_univariate_stacked_bar_chart(column):
    plt.figure(figsize=(8, 6))
    
    df_segmented = app_df_0[app_df_0['CODE_GENDER'].isin(['M', 'F', 'XNA'])]
    df_segmented[column].value_counts(normalize=True).plot(kind='bar', stacked=True, colormap='coolwarm')
    
    plt.title(f'Segmented Univariate Analysis - {column}')
    plt.xlabel(column)
    plt.ylabel('Proportion')
    plt.xticks(rotation=0)
    plt.tight_layout()
    
    # Save the chart as a PNG file
    plt.savefig(f'segmented_univariate_analysis_{column}.png')
    
    plt.show()

# Perform segmented univariate analysis for 'CODE_GENDER'
segmented_univariate_stacked_bar_chart('CODE_GENDER')
```

* We filter the data to include only the specified gender categories using .isin(['M', 'F', 'XNA']).
* We use the value_counts(normalize=True) function to calculate the proportion of each gender category.
* The stacked bar chart is created to visualize the proportions of each gender category.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bivariate Analysis - Side-by-Side Bar Chart
def bivariate_side_by_side_bar_chart(column):
    plt.figure(figsize=(8, 6))
    
    sns.countplot(data=app_df_0, x=column, hue='TARGET', palette='coolwarm')
    
    plt.title(f'Bivariate Analysis - {column} by TARGET')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.legend(title='TARGET', loc='upper right')
    plt.tight_layout()

     # Save the chart as a PNG file
    plt.savefig(f'bivariate_analysis_{column}.png')
    
    plt.show()

# Perform bivariate analysis for 'CODE_GENDER'
bivariate_side_by_side_bar_chart('CODE_GENDER')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting the columns related to amounts and the 'TARGET' variable:
columns_for_correlation = [
    'AMT_GOODS_PRICE',
    'AMT_ANNUITY',
    'AMT_CREDIT',
    'AMT_INCOME_TOTAL',
    'DAYS_BIRTH',
    'DAYS_EMPLOYED',
    'DAYS_REGISTRATION',
    'DAYS_LAST_PHONE_CHANGE',
    'DAYS_ID_PUBLISH',
    'TARGET'
]
df_columns_for_correlation = app_df_0[columns_for_correlation]

# Segment the dataset based on the target variable ('TARGET')
clients_with_payment_difficulties = df_columns_for_correlation[df_columns_for_correlation['TARGET'] == 1]
all_other_cases = df_columns_for_correlation[df_columns_for_correlation['TARGET'] == 0]

# Function to create a correlation matrix and heatmap
def create_correlation_heatmap(df_segmented, title, filename):
    plt.figure(figsize=(10, 8))
    corr_matrix = df_segmented.corr()
    cmap = sns.color_palette("coolwarm", as_cmap=True)
    sns.heatmap(corr_matrix, annot=True, cmap=cmap, fmt=".2f", center=0)
    plt.title(title)
    
    plt.show()

# Create correlation matrices and heatmaps for each scenario
create_correlation_heatmap(clients_with_payment_difficulties, 'Correlation Matrix - Clients with Payment Difficulties', 'clients_payment_difficulties_heatmap.png')
create_correlation_heatmap(all_other_cases, 'Correlation Matrix - All Other Cases', 'all_other_cases_heatmap.png')
```
